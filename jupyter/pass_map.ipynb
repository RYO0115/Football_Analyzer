```python
pass_data = [
    [20,20,50,30,1],
    [30,10,50,20,0],
    [40,-40,50,30,0],
    [-40,10,50,20,0],
    [-40,-20,50,-33,1],
    [-30,20,50,20,1],
    [-20,20,50,-30,0],
    [-30,10,50,10,1],
    [-20,0,-30,-10,1]
]
```

```python
pass_complete = dict(facecolor="red", width=1.0, headwidth=3.0, headlength=3.0, shrink=0.01)
pass_incomplete = dict(facecolor="blue", width=1.0, headwidth=3.0, headlength=3.0, shrink=0.01)
```

```python
import matplotlib.patches as mpatches

df = pd.DataFrame(pass_data,columns=["start_x","start_y","end_x","end_y","completed"])
```

```python
df
```

```python
completed_df = df[df["completed"]==1]
incompleted_df = df[df["completed"]==0]
```

```python
def DrawPassArrows( ax, df, color="red", scale=1):
    arrow = ax.quiver( df["start_x"], df["start_y"], df["end_x"]-df["start_x"], df["end_y"] - df["start_y"],color=color,scale=scale, angles="xy", scale_units="xy")
    return(arrow)
```

```python
completed_arrow = DrawPassArrows( ax, completed_df, color="red", scale=1.0)
imcompleted_arrow = DrawPassArrows( ax, incompleted_df, color="blue", scale=1.0)

ax.quiverkey(completed_arrow,X=0.3, Y=0.95, U=10,label="completed", labelpos="E", fontproperties={'size':13})
ax.quiverkey(imcompleted_arrow,X=0.7, Y=0.95, U=10,label="imcompleted", labelpos="E", fontproperties={'size':13})
    
```
